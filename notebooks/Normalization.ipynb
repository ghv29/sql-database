```python
import pandas as pd

df_job_company = pd.read_csv(r"C:\Users\rcrds\Desktop\Ironhack\Projects\Jobs SQL\sql-database\data\clean\To merge\Dataset1_Clean_Project4_GrpB_final.csv", sep=',')

df_job_company = df_job_company.drop(columns=["Rating","same_state","age","python_yn","R_yn","spark","aws","excel","Competitors","hourly","employer_provided","min_salary","max_salary","Salary Estimate","Company Name"])
df_job_company


```

```python
import re
df_job = df_job_company[["Job Title","Job Description","Location","avg_salary","job_state","company_txt"]]
df_job["Job Description"] = df_job["Job Description"].str.replace("\r\n","").str.replace("€™","").str.replace("â€","")
df_job = df_job.reset_index()
df_job = df_job.rename(columns={"index":"job_ID",
                                "Job Title":"title",
                                "Job Description":"description",
                                "avg_salary":"salary",
                                "job_state":"state",
                                "Location":"location"
                                })
def clean_for_mysql(text):
    if pd.isna(text):
        return text

    text = str(text)

    # 1. Remove ALL control characters (0x00–0x1F and 0x7F)
    text = re.sub(r"[\x00-\x1F\x7F]", " ", text)

    # 2. Normalize whitespace
    text = re.sub(r"\s+", " ", text)

    return text.strip()

df_job["description"] = df_job["description"].apply(clean_for_mysql)
df_job = df_job.rename(columns= {"company_txt":"Company Name"})
df_job["Company Name"] = df_job["Company Name"].str.replace("\r\n","")
```
