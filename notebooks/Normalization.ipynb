```python
import pandas as pd

df_job_company = pd.read_csv(r"C:\Users\rcrds\Desktop\Ironhack\Projects\Jobs SQL\sql-database\data\clean\To merge\Dataset1_Clean_Project4_GrpB_final.csv", sep=',')

df_job_company = df_job_company.drop(columns=["Rating","same_state","age","python_yn","R_yn","spark","aws","excel","Size","Competitors","hourly","employer_provided","min_salary","max_salary","Salary Estimate","Company Name","Headquarters"])
df_job_company


```

```python

df_job = df_job_company[["Job Title","Job Description","Location","avg_salary","job_state"]]
df_job["Job Description"] = df_job["Job Description"].str.replace("\r\n","")
df_job = df_job.reset_index()
df_job = df_job.rename(columns={"index":"job_ID",
                                "Job Title":"title",
                                "Job Description":"description",
                                "avg_salary":"salary",
                                "job_state":"state",
                                "Location":"location"
                                })
df_job
```

```python
df_job.to_csv(r"C:\Users\rcrds\Downloads\TABLE_JOB.csv",index=False)
```

```python
df_company = df_job_company[["Headquarters","Size","Founded","Type of ownership","Industry","Sector","Revenue","company_txt"]]
df_company = df_company.rename(columns= {"company_txt":"Company Name"})
df_company["Company Name"] = df_company["Company Name"].str.replace("\r\n","")
df_company = df_company.rename(columns={"Headquarters":"headquarters",
                           "Size":"size",
                           "Founded":"year_founded",
                           "Type of ownership":"ownership_type",
                           "Industry":"industry",
                           "Sector":"sector",
                           "Revenue":"revenue",
                           "Company Name":"name"})
df_company["year_founded"] = df_company["year_founded"].astype(int)
df_company
```

```python
df_isin = pd.read_csv(r"C:\Users\rcrds\Desktop\Ironhack\Projects\Jobs SQL\sql-database\data\clean\glassdoor_reviews\companiesISIN.csv")
df_isin = df_isin.drop_duplicates(subset="Company Name", keep="first")
df_isin.info()
```

```python
df_rating = pd.read_csv(r"C:\Users\rcrds\Desktop\Ironhack\Projects\Jobs SQL\sql-database\data\clean\To merge\Rating_Id.csv").drop_duplicates()
df_rating
```

```python
df_rating2 = pd.merge(df_rating, df_isin, how="left", left_on="company_isin", right_on="Company ISIN").drop(columns="Company ISIN")
df_rating2
```

```python
business_outlook_rating_avg = df_rating2["business_outlook_rating"].mean()
df_rating2["business_outlook_rating"] = df_rating2["business_outlook_rating"].fillna(business_outlook_rating_avg)
```

```python

ceo_rating_avg = df_rating2["ceo_rating"].mean()
df_rating2["ceo_rating"] = round(df_rating2["ceo_rating"].fillna(ceo_rating_avg),2)

compensation_benefits_rating_avg = df_rating2["compensation_benefits_rating"].mean()
df_rating2["compensation_benefits_rating"] = round(df_rating2["compensation_benefits_rating"].fillna(compensation_benefits_rating_avg),2)

culture_values_rating_avg = df_rating2["culture_values_rating"].mean()
df_rating2["culture_values_rating"] = round(df_rating2["culture_values_rating"].fillna(culture_values_rating_avg),2)

diversity_inclusion_rating_avg = df_rating2["diversity_inclusion_rating"].mean()
df_rating2["diversity_inclusion_rating"] = round(df_rating2["diversity_inclusion_rating"].fillna(diversity_inclusion_rating_avg),2)

recommend_friend_rating_avg = df_rating2["recommend_friend_rating"].mean()
df_rating2["recommend_friend_rating"] = round(df_rating2["recommend_friend_rating"].fillna(recommend_friend_rating_avg),2)

senior_management_rating_avg = df_rating2["senior_management_rating"].mean()
df_rating2["senior_management_rating"] = round(df_rating2["senior_management_rating"].fillna(senior_management_rating_avg),2)
df_rating2["overall_rating"] = round(df_rating2["overall_rating"],2)
df_rating2["business_outlook_rating"] = round(df_rating2["business_outlook_rating"],2)
df_rating2.isnull().sum()
```

```python
df_rating2 = df_rating2.rename(columns={"company_isin":"company_ISIN",
                           "business_outlook_rating":"business_outlook",
                           "ceo_rating":"ceo",
                           "compensation_benefits_rating":"compensation_benefits",
                           "culture_values_rating":"culture_values",
                           "diversity_inclusion_rating":"diversity_inclusion",
                           "number_of_reviews":"review_number",
                           "overall_rating":"overall",
                           "recommend_friend_rating":"recommend_friend",
                           "senior_management_rating":"senior_management"})
df_rating2 = df_rating2.drop(columns="Company Name")
df_rating2
```

```python
df_rating2.to_csv(r"C:\Users\rcrds\Downloads\TABLE_RATING.csv",index=False)
```

```python
def fillU(row):
    if pd.isnull(row["name"]):  # use pd.isnull for single values
        return row["Company Name"]
    else:
        return row["name"]
```
